```python
import pyautogui
import time

pyautogui.PAUSE = 0.5 #pausar 1 segundo a cada comando do pyautogui
link = "https://dlp.hashtagtreinamentos.com/python/intensivao/login"
pyautogui.press("win")
pyautogui.write("edge")
pyautogui.press("enter")
pyautogui.write(link)
pyautogui.press("enter")

time.sleep(3) #aguardar 3 segundos

pyautogui.hotkey("tab")
# pyautogui.click(x=674, y=447)
pyautogui.write("adaddas@gmail.com")
pyautogui.hotkey("tab")
pyautogui.write("awe213142rwfdas")
pyautogui.hotkey("tab")
pyautogui.press("enter")

time.sleep(3) #aguardar 3 segundos


# pip install pandas openpyxl   
import pandas

tabela = pandas.read_csv("produtos.csv")  #ou passar o caminho inteiro do arquivo
print(tabela)

# pandas.read_excel(sheet_name="Custos") #no ler arquivo excel consigo ler outra aba desta forma

#index linhas na tabela
#se fosse for coluna in tabela.columns: seria para cada coluna
#no python ele não conta a linha de cabeçalho, então ele pula a primeira linha
for linha in tabela.index:

    pyautogui.click(x=338, y=304)

    #codigo
    codigo = str(tabela.loc[linha, "codigo"]) #tabela do python sempre que for localizar info usar []
    pyautogui.write(codigo)
    # pyautogui.write("MOLO000251")
    pyautogui.press("tab")

    #marca
    marca = str(tabela.loc[linha, "marca"])
    pyautogui.write(marca)
    # pyautogui.write("MOLO000251")
    pyautogui.press("tab")

    #tipo
    tipo = str(tabela.loc[linha, "tipo"])
    pyautogui.write(tipo)
    # pyautogui.write("MOLO000251")
    pyautogui.press("tab")

    #categoria
    categoria = str(tabela.loc[linha, "categoria"])
    pyautogui.write(categoria) 
    # pyautogui.write("MOLO000251")
    pyautogui.press("tab")

    #preco
    preco = str(tabela.loc[linha, "preco_unitario"])
    pyautogui.write(preco)
    # pyautogui.write("preco")
    pyautogui.press("tab")

    #custo
    custo = str(tabela.loc[linha, "custo"])
    pyautogui.write(custo)
    # pyautogui.write("custo")
    pyautogui.press("tab")

    #obs
    obs = str(tabela.loc[linha, "obs"])
    if obs != "nan":
        pyautogui.write(obs)
    # pyautogui.write("obs")
    pyautogui.press("tab") #passou para botão enviar

    pyautogui.press("enter") #clicou em enviar

    #agora preciso voltar para o inicio da tela
    #poderia apertar home
    pyautogui.scroll(5000) #número positivo para cima, negativo para baixo

# str(3) #número 3 em formato de texto
```
